# **Estudando a Biblioteca SEABORN**

- Para o nosso estudo vamos utilizar um exemplo que se encontra dentro do Dataset Master da Biblioteca do Seaborn.

- Este exemplo contém o registro de alguns dados dos clientes que frequêntam um restaurante.

Arquivo: **tips**


```
# Importando a Biblioteca Seaborn
import seaborn as sns
```

```
# Carregar o arquivo que já se encontra no formato CSV
dados = sns.load_dataset("tips")

# Visualisando as 5 primeiras lindas do nosso DataFrame
dados.head()
```

### **Descrição das Variaveis do nosso DataFrame**

- **Total_bill** → *é o custo total de cada refição em US$*.

- **Tip** → *é o valor da luva ou gorjeta do garçom em US$*.

- **Sex** →  *é o tipo de Sexo das pessoas* (*Feminino / Masculino*).

- **Smoker** → *é o lugar ou a área escolhida dentro do restaurante* (se a pessoa é fumante ou não, pois existe lugares especificos).

- **Day** →  *Representa os dias da semana em que as pessoas frequêntam o restaurante*.

- **Time** →  *é tipo de refeição: Almoço (Lunch) ou Jantar (Dinner)*.

- **Size** → *é o tamanho do prato de refeição*.

```
# Verificando os tipos de dados usando o comando "INFO"
# É sempre bom fazer a verificação usando os dois comandos (DTYPES e INFO)
dados.info()
```

```
# Fazendo uma análise Matemática das Colunas numéricas
# Obtendo a "Média, Desvio Padrão, os valores Máximos e Mínimos e as variações percentuais em tornda média"
dados.describe()
```

```
# Visualisndo o valor total a ser pago e o valor da luva pelo atendimento
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", data=tips)
```

**Visualualização**


- O **replote** é considerado um método de alto nível quando se busca uma boa visualização de dados fazendo uma relações estatísticas entre as variáveis em análise, pois ele combina o método **scatter plot** e **line plots**.

- *O parâmetro **hue** é usado par informar em qual coluna será  colorida* (no nosso caso, quais pontos serão coloridos).

```
# Visualindo de forma clara e estatística
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

**Conclusão da análise feita nos gráficos acima**

- *Nos gráficos de dispersão plotados pode-se observar que existe uma coerência com o resultado esperado, pois quanto maior o valor da refeição maior é o valor da luva, com isso podemos cocluir que existe uma boa relação entre as duas variavési em análise*.

```
# Usando outra variável para colorir
sns.relplot(x="total_bill", y="tip", hue="sex", data=tips)
```

```
# Mudando o estilo da variável "Sex" e colorindo ela para que possamos visualizar
# Note agora que temos marcadores diferentes para o tipo de sexo (F/M)
# Assim podemos enteder que se tivermos analisando uma grandeza com mais tipos ele vai colocar diferentes pontos.

sns.relplot(x="total_bill", y="tip", hue="sex", style="sex", data=tips)

# Observa que agora é muito mais facíl diferencias a quantidade de pessoas olhando pelo sexo
```

```
# Essa representação pode ser feita usando 4 variáveis com tonalidades diferentes
sns.relplot(x="total_bill", y="tip", hue="sex", style="smoker", data=tips)

# Note que temos mais de 3 variáveis no mesmo gráfico, e o gráfico fica muito poluído dificultando a extração das informações.
```

```
# Fazendo o mesmo procedimento usando variáveis numerica
sns.relplot(x="total_bill", y="tip", hue="size", data=tips)

# Note que temos mais de 3 variáveis no mesmo gráfico, e o gráfico fica muito poluído dificultando a extração das informações.
```

```
# Refazendo o mesmo processo, porem usando uma única cor
sns.relplot(x="total_bill", y="tip", size="size", data=tips)

# Note que temos mais de 3 variáveis no mesmo gráfico, e o gráfico fica muito poluído dificultando a extração das informações.
```

```
# Alterando o tamnho dos pontos
# Agora é possível perber com clareza os pontos maiores e menores que representam o tamanho da refeição
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15,300), data=tips)
```

## Agora vamos usar o **Scatterplot**

- Vale lembrar que o **Scatterplot** também é uma forma de plotar os gráficos de dispersão, porém ele só plota gráfico de dispersão, ou seja, não contém a função **lineplot()**.

```
# Analisando a relação entre as 4 variáveis "total_bill, tipo, day, smoker"
sns.scatterplot(x="total_bill", y="tip", hue="day", style="time", data=tips)
# Mas uma vez podemos observar mais de 3 variáveis no mesmo gráfico, perde-se a qaulidade e 
# a visibilidade mesmo com as legendas.
# Com isso o gráfico fica muito poluído dificultando a extração das informações.
```

```
# Podemos redimensionar as figuras no "Seaborn" pois ele herda algumas funçãoes do "Matplotlib"
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(15, 7))
sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips)
```

# ***Visualizando e analisando as Cores do Seaborn***

##                *Temos várias funções de cores dentro do Seaborn, irei destacar apenas algumas.*


- O **color_palette()** → *ele faz um controle das cores e também recebe o nome de paleta de cores dentro do seabor*. 


- O **set_palette()** → esse parâmetros é usado para definir um estilo padrão de cores para todos os gráficos que for plotar.

Lembre-se que ao definir uma cor deve-se levar em contas o tipo de dado que queremos analisar para que assim posso plotar o gráfico. Os dados a serem plotados eles podem ser:

- *Qualitativos*

- *Sequênciais*

- *Divergentes*



```
# Verificando e Plotando as Cores do tipo QUALITATIVAS
sns.palplot(sns.color_palette("Accent"))
```

```
# Verificando e Plotando as Cores do tipo DIVERGENTES
sns.palplot(sns.color_palette("PuOr"))
```

```
# Verificando e Plotando as Cores do tipo SEQUÊNCIAIS
sns.palplot(sns.color_palette("Blues"))
```

```
# Vamos plotar um gráfico de dispersão usando a paleta SEQUÊNCIAL
# Com isso é possível perceber que temos uma visualização harmônica dos pontos, pois, bem como 
# a intensidade de cada um dos pontos presentes.

sns.relplot(x="total_bill", y="tip", hue="size", palette="Blues", size="size", sizes=(12,250), data=tips)
```

```
# Gráfico usando a paleta de cor "QUALITATIVA" “Set2”. Também chamdo de "GRÁFICO DE BARRAS"
# Você pode testar usano o "SET1" para verificar como ficrá o seu gráfico.

sns.catplot(x="sex", kind="count", palette="Set2", data=tips)

# Note que não existe uma variação das cores, pois temos dados categóricos e não contínuos como os da coluna size.
```

```
# Analisando e Plotando gráfico usando a paleta de cores DIVERGENTES “PiYG”.

# As paletas também são excelentess quando trabalhamos com "MATRIZES de CORRELAÇÃO", pois a tonalidade da coloração
# identifica o grau de correlação entre as variáveis em análise.

sns.heatmap(tips.corr(), annot=True, cmap="PiYG")
```

```
# Agora vamos plotar gráficos definindo uma cor PADRÃO usando a paleta de cor "PASTEL"

sns.set_palette("pastel")
sns.relplot(x="total_bill", y="tip", data=tips)
```

```
# Plotando o gráfico definindo uma cor TEMPORÁRIA
with sns.color_palette("Accent"):
    sns.relplot(x="total_bill", y="tip", size="size", sizes=(12,250), data=tips)
```

# Vamos analisar e plotar gráfico dos dados **CATEGÓRICOS**

- Quando estamos trabalhando com as variáveis categóricas é preciso entender que visualização dos dados se dá de forma diferente.

- Na biblioteca **Seaborn** encontramos o método **catplot()** que possui diversos tipos de gráficos dentro dele para se fazer este tipo de análise, com isso fica muito mais simples fazer diferentes gráficos com o mesmo método.

```
# Vamos plotar um gráfico de Barras (também conhecido como gráfico de colunas)
# Note que também podemos identificar os eixo, e a quantidade de registros feito de forma automatica
sns.catplot(x="sex", kind="count", palette="Set2", data=tips)
```

```
# Vamos plotar um gráfico de BARRA HORIZONTAL (também conhecido como gráfico de colunas Horizontal)
# Note que também podemos identificar os eixo, e a quantidade de registros feito de forma automatica
# Aqui informamos que os dados estão no eixo y e o número de registos agora é mostrado no eixo x
sns.catplot(y="day", kind="count", palette="Set1", data=tips)
```

```
# Plotando o gráfico usando o "SCATTER PLOT"

sns.catplot(x="day", y="total_bill", palette="Set2", data=tips)
```

```
# Plotando o gráfico usando o parâmetro "SWARM" que serve para evita sobreposição de pontos
sns.catplot(x="day", y="total_bill", kind='swarm', data=tips)

# Note que não foi preciso informamos o tipo de gráfico a ser plotado.
# Com isso dá pra visualizar como está a distribuição das refeições vendidas pelo dia da semana, onde cada "COR" representa um dia.
```

```
# Conforme já vimos anteriormente vamos usar o parâmetro "HUE" para adicionarmos uma terceira variável a nossa visualização
sns.catplot(x="day", y="total_bill", kind='swarm', hue="sex", palette="Dark2", data=tips)
```

```
# Plotando o gráfico usando o parâmetro "ORDE" para alterarmos a ordem padrão das categorias que estão sendo exibidas.
# Isso é importante quando temos mais de um gráfico na mesma figura e queremos mudar ou manter a mesma ordem.

sns.catplot(x="day", y="total_bill", kind='swarm', hue="sex", order=['Sat','Sun','Fri','Thur'], data=tips)
```

```
# Colocanda os gráficos na horizontal
sns.catplot(x="total_bill", y="day", kind="swarm", data=tips)
```

# **Plotando gráficos com Regressão**

- Quando se tem muitas variáveis do tipo **quantitativas** nos dados em análise é interessante fazer uma visualização para sabermos como elas se relacionam, pois isto ajuda muito na interpretação dos dados e dos resultados.

- E essa visualização pode ser feita usando linhas de regressão.

- Com modelos exisitentes sobre de **regressão simples** é possível verificar se existe alguma correlação entre algumas variáveis em estudo. 

```
#  Nesses pontos, a linha azul representam o sexo feminino e a linha rosa o sexo masculino.
# Vamos verificar se existe uma correlação positiva entre as variáveis e se os pontos se estão mais aproximos da linha de regressão.

# IMPORTANTE: Esse gráfico não representa uma regressão do nosso conjunto de dados, apenas estamos mostrando 
# que existe uma relação entre esses dados.

%matplotlib inline
sns.lmplot(x="total_bill", y="tip", hue="sex", palette="Pastel1", data=tips)
```

```
# Plotando gráfico usando o parâmetro "COL" que serve para separar os gráficos pelo valor da coluna "TIME".

sns.lmplot(x="total_bill",y="tip", hue="smoker", col="time", palette="Set2", data=tips)

# Observa que na coluna "TIME" tem-se 2 valores diferentes que são:  Lunch e Dinner 
# Assim, é possível segmentar os gráficos de correlação por coluna.
```

```
# Plotando o mesmo gráfico.
# Dando mais enfas nas linhas de regresão com os parâmetros "SCATTER_KWS e LINE_KWS" .
sns.lmplot(x="total_bill",
                  y="tip", 
                 hue="smoker", 
                 col="time", 
                 palette="Set1",
                 data=tips, 
                 scatter_kws={'s': 90, 'alpha': 0.5},
                 line_kws={'lw': 4});
```

```
# Vamos separar os gráficos usando os parâmetros "COL e ROW".

sns.lmplot(x="total_bill", y="tip", hue="smoker",col="time", row="sex", palette="Set2", data=tips)

# Vamos fazer essa separação por horário de refeição e pelo sexo.
# Lembre-se que o "COL" é a segmentação por coluna e o "ROW" é a segmentação por linha.
```

# ***Visualização Outliers dos dados***

## *Gráficos BoxPlot*

```
# Plotando um gráfico dos dias da semana em relação ao custo total
sns.catplot(x="day", y="total_bill", kind="box", data=tips)

# Note que para cada dia da semana é possível ver o limite inferior e superior, os quartis, a mediana e como os dados estão distribuídos. 
```

```
# Plotando o gráfico de Boxplot dos dias pelo total de conta e pelo vericando a área em que a pessoa se encontra.
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips)

# IMPORTANTE: Plotar esse tipo de gráfico é interessante quando se trabalha com uma grande quantidade de dados.
```

```
# Plotando gráficos usando o "Método STRIPPLOT" que permite plotar a distribuição dos dados de uma outra forma.
sns.catplot(x="day", y="total_bill", kind="boxen", data=tips)
```

```
# Combinando os dois gráficos

sns.catplot(x="day", y="total_bill", kind="boxen", data=tips)
sns.stripplot(x="day", y="total_bill", data=tips, color="blue")

# Entenda que os pontos mostram como está a distribuição dos dados dentro de cada boxplot.
# IMPORTANTE: Para uma grande quantidade de dados essa combinação não é indicada, porque deixa a visualização muito poluída.
# IMPORTANTE: Isso não é recomendavel fazer.
```

```
# O "CATPLOT()" é um método que permite usarmos "FACETGRID"
# "FACETGRID" ->> é uma classe que cria uma matriz/grid em branco que podemos preencher com qualquer gráfico
# Com as "FACETGRID" podemos combinar gráficos em uma única figura.
%matplotlib inline
sns.catplot(x="sex", y="total_bill", hue="smoker", col="time", data=tips, kind="boxen",height=4);
```

# ***Plotando Gráficos de Correlação***

- Esses tipo de gráficos são úteis quando queremos visualizar se existe correlação positivas ou negativas entre colunas em análise.

- O método **corr()** da biblioteca *pandas* nos possibilita determinar a correlação por cada um dos tipos

    ### *Esses tipos podem ser:*

- **spearman** 
-  **pearson**.

```
# Verificando a correlação das colunas numericas
# IMPORTANTE: As correlações só podem ser calculadas ou feitas para as colunas numéricas.
correlacoes = tips.corr()
correlacoes
```

```
# Verificando a correlação através das cores
# IMPORTANTE: Quanto mais claro a cor for, maior a correlação e quanto mais escura a cor for, menor é a correlação.
ax = sns.heatmap(correlacoes)
```

```
# Extraindo e exibindo os valores da correlação
ax = sns.heatmap(correlacoes, annot=True)
```
